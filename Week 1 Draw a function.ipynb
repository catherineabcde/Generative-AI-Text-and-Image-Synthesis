## **<font color="#0000FF">Áï´‰∏ÄÂÄãÂáΩÊï∏ÂúñÂΩ¢ÔºàWeek 1 ‰ΩúÊ•≠Ôºâ</font>**

‰ª•‰∏ãÂúñÂΩ¢‰æùÊìöË≥áÊñô‰æÜÊ∫êÁ®±ÁÇ∫Á≠âËßíËû∫Á∑öÔºåÊòØÈòøÂü∫Á±≥Âæ∑ÁöÑÁ≠âÈÄüËû∫Á∑öÁöÑÊõ¥ÈÄ≤ÂåñÁâàÔºÅ
Âú®Ê•µÂ∫ßÊ®ô $(r, \theta)$ ‰∏≠ÔºåÊ≠§Êõ≤Á∑öÂèØ‰ª•ÂØ´ÁÇ∫Ôºö

$$r = a*e^{b\theta}$$



> üìÇ Ë≥áÊñô‰æÜÊ∫êÔºö[Á≠âËßíËû∫Á∑ö](https://zh.wikipedia.org/zh-tw/%E7%AD%89%E8%A7%92%E8%9E%BA%E7%BA%BF)

###  **<font color="#0000FF">ÈÅ∏ÊìáÊ≠§ÂáΩÊï∏ÂúñÂΩ¢ÁöÑÂéüÂõ†</font>**

Âú®ÈÅ∏ÊìáÂúñÂΩ¢ÁöÑÈÅéÁ®ãÁúãÂà∞Ë®±Â§öÊºÇ‰∫ÆÁöÑÊï∏Â≠∏ÂáΩÊï∏Ôºå‰ΩÜÊòØÂæàÂ§öÈÉΩÊòØÁÇ∫‰∫ÜÈÄ†Â∞±ÈÇ£ÂÄãÂΩ¢ÁãÄËÄåÊ≤íÊúâÁâπÂà•Áâ©ÁêÜÊÑèÁæ©ÁöÑÂúñÂΩ¢„ÄÇÁï∂ÊàëË™çË≠òÂà∞ÈÄôÂÄãÂáΩÊï∏ÂúñÂΩ¢ÊôÇÔºåÈô§‰∫ÜÁ≠âËßíÈÄüÂ∫¶ÁöÑÁâπÊÄßÈÇÑÊúâËÉåÂæåÁöÑÊ≠∑Âè≤ÊïÖ‰∫ãÈÉΩÂæàÊúâË∂£ÔºåÊâÄ‰ª•ÊúÄÂæåÂ∞±Ê±∫ÂÆöÈÅ∏ÊìáÂÆÉ‰∫Ü„ÄÇ

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = 0.1
b = 0.05
turns = 200
n = 10000

theta = np.linspace(0.0, 2*turns*np.pi, n)
r = a*np.exp(b*theta)

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x, y, color="#998dff", marker="*", linewidth=4) # color: ÊéßÂà∂È°èËâ≤; marker: ÊéßÂà∂ÈªûÁöÑÂΩ¢ÁãÄ; linewidth: ÊéßÂà∂ÂúñÂΩ¢ÁöÑÁ∑öÊ¢ùÂØ¨Â∫¶
```

*   `a`: Ëû∫Á∑öÁöÑËµ∑ÂßãÂçäÂæëÔºåÊéßÂà∂ÁæÖÁ∑öÊï¥È´îÂúñÂΩ¢ÁöÑÂ§ßÂ∞è„ÄÇ
*   `b`: Ê±∫ÂÆöÂçäÂæëÈö®ËßíÂ∫¶Â¢ûÈï∑ÁöÑÈÄüÂ∫¶ÔºåËàáÁ≠âËßíËßíÂ∫¶ÊúâÁõ¥Êé•Èóú‰øÇÔºåÂèØ‰ª•Ë™™ÊòØÈñãÂè£ËßíÂ∫¶ÁöÑÊéßÂà∂Âô®ÔºåË∂äÂ§ßË∂äÁ∑äÂØÜÔºåË∂äÂ∞èË∂äÈñãÊîæ„ÄÇ
*   `turns`: Ëû∫Á∑öËµ∞ÁöÑÁØÑÂúçÔºàËû∫Á∑öÊóãËΩâÂúàÊï∏Ôºâ„ÄÇ
*   `n`: ÈÅ∏ÂèñÊ®£Êú¨Êï∏ÈáèÔºàÂêåËÄÅÂ∏´ÁöÑ‰∏äË™≤ÁØÑ‰æãÔºâÔºåÂ¶ÇÊûú `turns` Êàñ `n` Êï∏ÈáèÊ∏õÂ∞ëÂúñÂΩ¢ÊúÉËÆäÊàêÈã∏ÈΩíÁãÄ„ÄÇ
*   `theta`: Â∏åÊúõÂúñÂΩ¢Ê∂µËìãÁöÑËßíÂ∫¶ÁØÑÂúçÔºåÂ¶ÇÊûúÁ∏ÆÂ∞èË®≠ÂÆöÁØÑÂúçÔºåÊâÄÂëàÁèæÁöÑÂúàÊï∏Â∞±ÊúÉËÆäÊØîËºÉÂ∞ëÂúàÊàñÊòØÁÑ°Ê≥ïÂª∫ÊßãÂÆåÊï¥ÂúìÂΩ¢„ÄÇ
